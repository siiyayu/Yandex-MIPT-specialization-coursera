### Выберем распределение хи-квадрат: для него mean = k, variance = 2*k

```python
import scipy.stats as sts
import numpy as np
import matplotlib.pyplot as plt
```

### Создаем матрицу выборок:

```python
distr = sts.chi2(3) #количество степеней свободы возьмем 3
means = np.zeros((3, 1000)) #создаем пустую матрицу для хранения средних значений
size = [5, 10, 50]  #список объемов выборок
d = np.zeros((1, 1000))
for n in size:
    for i in range(1000):
        means[size.index(n), i] = np.mean(distr.rvs(n))        
```

```python
means #посмотрим, что у нас получилось
```

### Построение графиков и накладывание на них гауссовских распределений:

#### n = 5

```python
plt.hist(means[0], bins = 20, density = True) # n = 5
plt.plot()
plt.ylabel('$fraction of samples$')
plt.xlabel('$means$')
norm_dist = sts.norm(3, np.sqrt(2*3/size[0]))
x = np.linspace(0, 8, 200)
plt.plot(x, norm_dist.pdf(x))
```

#### n = 10

```python
plt.hist(means[1], bins = 20, density = True) # n = 10
plt.ylabel('$fraction of samples$')
plt.xlabel('$means$')
norm_dist = sts.norm(3, np.sqrt(2*3/size[1]))
x = np.linspace(0, 8, 200)
plt.plot(x, norm_dist.pdf(x))
```

#### n = 50

```python
plt.hist(means[2], bins = 20, density = True) # n = 50
plt.ylabel('$fraction of samples$')
plt.xlabel('$means$')
norm_dist = sts.norm(3, np.sqrt(2*3/size[2]))
x = np.linspace(1, 5, 200)
plt.plot(x, norm_dist.pdf(x))
```

### Вывод:

Аппроксимация выборочных средних гауссовским распределением улучшается при увеличении размера выборки
