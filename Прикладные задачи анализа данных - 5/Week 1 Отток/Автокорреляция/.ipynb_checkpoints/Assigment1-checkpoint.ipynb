### Уровень заработной платы в Росии

```python
%pylab inline
import pandas as pd
from scipy import stats
import statsmodels.api as sm
import matplotlib.pyplot as plt
import warnings
from itertools import product

def invboxcox(y,lmbda):
   if lmbda == 0:
      return(np.exp(y))
   else:
      return(np.exp(np.log(lmbda*y+1)/lmbda))
```

В файле записаны данные о заработной плате за каждый месяц с января 1993 по август 2016. .

Необходимо проанализировать данные, подобрать для них оптимальную прогнозирующую модель в классе ARIMA и построить прогноз на каждый месяц на два года вперёд от конца данных.

Для начала загрузим данные с помощью pandas.


```python
data = pd.read_csv("WAG_C_M.csv",';', index_col=['month'], 
                   parse_dates=['month'], dayfirst=True)
data.head()
```

### Визуальный анализ ряда

```python
plt.figure(figsize(15,7))
data.WAG_C_M.plot()
plt.ylabel('Salaries')
pylab.show()
```

На графике четко прорисовывается возростающий тренд и сезонность. Ряд нестационарен. Чтобы в этом убедиться можем также переверить за критерием Дики-Фуллера.

Проверка стационарности и STL-декомпозиция ряда:

```python
plt.figure(figsize(15,10))
sm.tsa.seasonal_decompose(data.WAG_C_M).plot()
print("Критерий Дики-Фуллера: p=%f" % sm.tsa.stattools.adfuller(data.WAG_C_M)[1])
```

С STL-декомпозиции ряда видно, что тренд действительно ярко выраженый возрастающий, присутствует годовая сезонность.

Со временем дисперсия ряда изменяеться, проведем её стабилизацию.

### Стабилизация дисперсии

Стабилизацию дисперсии можно сделать благодаря преобращованию Бокса-Кокса.

```python
data['WAG_C_M_box'], lmbda = stats.boxcox(data.WAG_C_M)
plt.figure(figsize(15,7))
data.WAG_C_M_box.plot()
plt.ylabel(u'Transformed salaries')
print("Оптимальный параметр преобразования Бокса-Кокса: %f" % lmbda)
print("Критерий Дики-Фуллера: p=%f" % sm.tsa.stattools.adfuller(data.WAG_C_M_box)[1])
```

### Стационарность и выбор порядка дифференцирования

После стабилизации дисперсии, ряд все еще нестационарен, что видно как с критерия Дики-Фуллера, так и с графика. Проведем сезонное дифференцирование. На полученом ряде проверим стационарность и посмотрим на STL-декомпозицию полученого ряда.

```python
data['WAG_C_M_box_diff'] = data.WAG_C_M_box - data.WAG_C_M_box.shift(12)
plt.figure(figsize(15,10))
sm.tsa.seasonal_decompose(data.WAG_C_M_box_diff[13:]).plot()
print("Критерий Дики-Фуллера: p=%f" % sm.tsa.stattools.adfuller(data.WAG_C_M_box_diff[12:])[1])
```

Гипотеза нестационарности отвергается за критерием Дики-Фуллера. Но визуально тренд еще видно, можем использовать также обычное диференцирование.

```python
data['WAG_C_M_box_diff1'] = data.WAG_C_M_box_diff - data.WAG_C_M_box_diff.shift(1)
plt.figure(figsize(15,10))
sm.tsa.seasonal_decompose(data.WAG_C_M_box_diff1[13:]).plot()
print("Критерий Дики-Фуллера: p=%f" % sm.tsa.stattools.adfuller(data.WAG_C_M_box_diff1[12:].dropna())[1])
```

```python
data.WAG_C_M_box_diff1.plot()
```

Теперь от от тренда избавились и можна строить модель

## Подбор модели

#### Выбор начальных приближений для p,q,P,Q

Посмотрим на ACF и PACF ряда и попробуем с их помощью выбрать начальные приближения параметров.

```python
plt.figure(figsize(15,8))
ax = plt.subplot(211)
sm.graphics.tsa.plot_acf(data.WAG_C_M_box_diff1[13:].values.squeeze(), lags=48, ax=ax)
pylab.show()
ax = plt.subplot(212)
sm.graphics.tsa.plot_pacf(data.WAG_C_M_box_diff1[13:].values.squeeze(), lags=48, ax=ax)
pylab.show()
```

На графике аутокореляции не прослеживается сезонность, значит Q возьмем равным 0.

Последний значимый несезонный лаг на этом графике равный 11, возьмем q = 1.

На графике частичной аутокорреляции также нет никакой сезонности, поэтому выберем P = 0. 

Последний значимый лаг здесь в единице, берем p = 1.

```python
ps = range(0, 2)
d=1
qs = range(0, 2)
Ps = range(0, 1)
D=1
Qs = range(0, 1)
```

#### Обучение и сравнение моделей-кандидатов, выбор победителя

```python
parameters = product(ps, qs, Ps, Qs)
parameters_list = list(parameters)
len(parameters_list)
```

Посмотрим есть ли параметры на которых модель не обучается и выведем их.

```python
%%time
results = []
best_aic = float("inf")
warnings.filterwarnings('ignore')

for param in parameters_list:
    #try except нужен, потому что на некоторых наборах параметров модель не обучается
    try:
        model=sm.tsa.statespace.SARIMAX(data.WAG_C_M_box, order=(param[0], d, param[1]), 
                                        seasonal_order=(param[2], D, param[3], 12)).fit(disp=-1)
    #выводим параметры, на которых модель не обучается и переходим к следующему набору
    except ValueError:
        print('wrong parameters:', param)
        continue
    aic = model.aic
    #сохраняем лучшую модель, aic, параметры
    if aic < best_aic:
        best_model = model
        best_aic = aic
        best_param = param
    results.append([param, model.aic])
    
warnings.filterwarnings('default')
```

Таких пареметров не оказалось. Теперь можна посмотреть на лучшие 5 моделей

```python
result_table = pd.DataFrame(results)
result_table.columns = ['parameters', 'aic']
print(result_table.sort_values(by = 'aic', ascending=True).head())
```

Наилучшей оказалась модель с p = 1, q = 10, Q = 0 и P = 0.

Выведем на екран информацию об этой моделе.

```python
print(best_model.summary())
```

#### Анализ остатков построенной модели

```python
plt.figure(figsize(15,8))
plt.subplot(211)
best_model.resid[13:].plot()
plt.ylabel(u'Residuals')

ax = plt.subplot(212)
sm.graphics.tsa.plot_acf(best_model.resid[13:].values.squeeze(), lags=48, ax=ax)

print("Критерий Стьюдента: p=%f" % stats.ttest_1samp(best_model.resid[13:], 0)[1])
print("Критерий Дики-Фуллера: p=%f" % sm.tsa.stattools.adfuller(best_model.resid[13:])[1])
```

С критерия Стьюдента видим, что остатки несмещены и за критерием Дики-Фуллера стационарны. Также с коррелограммы видно что остатки неавтокоррелированы. Модель не нуждаеться в дополнительной коректировке

Можем посмотреть, как хорошо модель описывает реальные данные:

```python
data['model'] = invboxcox(best_model.fittedvalues, lmbda)
plt.figure(figsize(15,7))
data.WAG_C_M.plot()
data.model[13:].plot(color='r')
plt.ylabel('Wine sales')
pylab.show()
```

Видим, что данные хорошо описываються этой моделью.

И в завершение проведем предсказание на следущие три года.

```python
data2 = data[['WAG_C_M']]
date_list = [datetime.datetime.strptime("2016-09-01", "%Y-%m-%d") + relativedelta(months=x) for x in range(0,36)]
future = pd.DataFrame(index=date_list, columns= data2.columns)
data2 = pd.concat([data2, future])
data2['forecast'] = invboxcox(best_model.predict(start=284, end=320), lmbda)

plt.figure(figsize(15,7))
data2.WAG_C_M.plot()
data2.forecast.plot(color='r')
plt.ylabel('Salaries')
pylab.show()
```

Прогноз выглядит вполне адекватно
