Удалим строки согласно условию задания. 
    - нулевой длительностью
    - нулевым количеством пассажиров
    - нулевым расстоянием поездки по счётчику
    - координатами начала, не попадающими в прямоугольник Нью-Йорка

```python
data = data.loc[(data['passenger_count']!= 0) & 
                    (data['trip_distance']!= 0) & 
                    (data['tpep_pickup_datetime']!=data['tpep_dropoff_datetime'])
                  &(((data['pickup_longitude'] > New_York_rect[0][0]) & (data['pickup_longitude'] < New_York_rect[0][1]))
                &((data['pickup_latitude'] > New_York_rect[1][0]) & (data['pickup_latitude'] < New_York_rect[1][1]) ))]
```

```python
data.shape # удалили почти 200к записей 
```

Разделим атрибут "tpep_pickup_datetime" на дату и часы, чтобы было возможно аггрегировать и заполнять пропущенные пары (часы, область)

```python
def get_int_value(ts):
    if ts[0]=='0':
        return int(ts[1])
    else:
        return int(ts)
```

```python
data['hour'] = data['tpep_pickup_datetime'].apply(lambda x: get_int_value((str(x).split(" ")[1]).split(":")[0]))
```

```python
data['day'] = data['tpep_pickup_datetime'].apply(lambda x: get_int_value(str(x).split(" ")[0].split("-")[2]))
```

```python
data
```

```python
def cord_to_index(data1 , data2): #функция принимает 2 серии с координатами и возвращает индекс региона
    w = (New_York_rect[0][1] - New_York_rect[0][0]) / 50 # ширина области 
    h = (New_York_rect[1][1] - New_York_rect[1][0]) / 50 # высота области 
    ind_w = (data1 - New_York_rect[0][0]) // w   #номер по ширие т.е ячейка от 0 до 49  
    ind_h = (data2 - New_York_rect[1][0]) // h   #номер по высоте т.е ячейка от 0 до 49
    #нумерация в файле regions происходит  слево-направо , снизу - вверх , 
    return 50*ind_w + ind_h + 1
```

```python
data['region_id']= cord_to_index(data["pickup_longitude"] , data["pickup_latitude"]).astype('int')
```

```python
data_count = data.pivot_table(values =['trip_distance'] , index = ['date','hour'] , columns = ['region_id'] , aggfunc = 'count' , fill_value = 0 , margins_name = 'count')
```

```python
data_count = data_count.rename(columns = {'trip_distance':'count'})
```

Соберем все в одну функцию, которая будет принимать имя файла и создавать итоговые данные. В дальнейшем мы будем применять ее для обработки всех входных данных.

```python
import os 
```

```python
def prepare_data(filename , src_dir,  des_dir):
    data = pd.read_csv(os.path.join(src_dir,filename) , usecols = range(1,7))
    New_York_rect = ((-74.25559,-73.70001), (40.49612,40.91553))
    
    data = data.loc[(data['passenger_count']!= 0) & 
                    (data['trip_distance']!= 0) & 
                    (data['tpep_pickup_datetime']!=data['tpep_dropoff_datetime'])
                  &(((data['pickup_longitude'] > New_York_rect[0][0]) & (data['pickup_longitude'] < New_York_rect[0][1]))
                &((data['pickup_latitude'] > New_York_rect[1][0]) & (data['pickup_latitude'] < New_York_rect[1][1]) ))]
    
    data['hour'] = data['tpep_pickup_datetime'].apply(lambda x: get_int_value((str(x).split(" ")[1]).split(":")[0]))
    data['day'] = data['tpep_pickup_datetime'].apply(lambda x: get_int_value(str(x).split(" ")[0].split("-")[2]))
    data['region_id']= cord_to_index(data["pickup_longitude"] , data["pickup_latitude"]).astype('int')
    
    data_count = data.pivot_table(values =['trip_distance'] , index = ['day','hour'] , columns = ['region_id'] , \
                                  aggfunc = 'count' , fill_value = 0)
    
    data_count = data_count.rename(columns = {'trip_distance':'count'})
    for k in range(1,2501):
        if ('count',k) not in data_count.columns:
            data_count[('count',k)] = pd.Series(np.zeros(data_count.shape[0],))
            data_count[('count',k)] = data_count[('count',k)].fillna(0).astype('int')
    
    data_count = data_count.reindex(sorted(data_count.columns), axis=1)
    
    data_count.to_csv(os.path.join(des_dir ,filename) , index = False , header = False)
    
```

```python
src_dir  = "yellow_taxi_data" 
filename = "yellow_tripdata_2016-05.csv"
```

```python
prepare_data(filename , src_dir, 'timed_data')
```
